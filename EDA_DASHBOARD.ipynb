# **REMOVING OUTLIERS USING NORML DISTRIBUTION(GAUSSIAN DISTRIBUTION)** 

```python
from scipy.stats import norm
import numpy as np

plt.hist(df.AMT_INCOME_TOTAL, bins=20, rwidth=0.8, density=True)
plt.xlabel('TOTAL INCOME')
plt.ylabel('Count')

rng = np.arange(-5000, df.AMT_INCOME_TOTAL.max(), 100)
plt.plot(rng, norm.pdf(rng,df.AMT_INCOME_TOTAL.mean(),df.AMT_INCOME_TOTAL.std()))
```

MEAN OF AMT_INCOME_TOTAL

```python
mean = df.AMT_INCOME_TOTAL.mean()
mean
```

STANDARD DEVIATION OF AMT_INCOME_TOTAL

```python
st_deviation = df.AMT_INCOME_TOTAL.std()
st_deviation
```

3th STANDARD DEVIATION OF AMT_INCOME_TOTAL

```python
mean + 3* st_deviation
```

```python
mean - 3* st_deviation
```
